# Extract info from list of days and save it as json

```python
import json
global_contents = None
test_split = {}
test_split["DAYS"] = []

with open("days.txt", "r") as input_file:
    contents = input_file.readlines()
    global_contents = contents
    # print(contents)


number_of_lines = len(global_contents)
print(global_contents[0])

for index in range(0,number_of_lines):
    
    split_line = global_contents[index].split("-",2)
    # print(f"Day:{index + 1} | Language:{split_line[1].strip()} | Description: {split_line[2].strip()}")
    test_split["DAYS"].append(
        {
            'Day': index + 1,
            'Tech': split_line[1].strip(),
            'description': split_line[2].strip()
        }
    )

json_string = json.dumps(test_split,indent=2)
print(json_string)

with open("codeeveryday-days.json", "w") as writeJson:
    json.dump(test_split, writeJson, indent=2)
```
